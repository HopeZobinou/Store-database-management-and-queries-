```python

sql_check = "SELECT cust_id from tCust WHERE first LIKE ? AND last LIKE ? AND addr LIKE ? AND zip LIKE ?;"
sql_insert = "INSERT INTO tOrder (cust_id, date) VALUES (?, ?);"

for i, row in enumerate(tOrder.to_dict(orient = 'records')):
    order_params = (f"%{row['first']}%", f"%{row['last']}%", f"%{row['addr']}%", f"%{row['zip']}%") 
    curs.execute(sql_check, order_params)

    result = curs.fetchone()
    #cust_id = pd.read_sql(sql_check, conn, params = order_params)

    if result:
        cust_id = result[0]
        try:
            curs.execute(sql_insert, (cust_id, row['date']))
        except Exception as e:
            print(e)
            print('\nrow: ', i)
            print('\n', row)
            break
```

```python
curs.execute("DROP TABLE IF EXISTS tOrderDetail;")
sql = """
CREATE TABLE tOrderDetail (
    order_id INTEGER REFERENCES tOrder(order_id),
    prod_id INTEGER REFERENCES tProd(prod_id),
    qty INTEGER NOT NULL,
    PRIMARY KEY(order_id, prod_id)
)
;"""
curs.execute(sql)
```

```python
tOrderDetail = store_data_all_months[['first', 'last', 'addr', 'zip', 'prod_id', 'qty','date']]

tOrderDetail
```

```python
sql_check = """SELECT order_id from tOrder 
                JOIN tCust USING(cust_id)
                WHERE first LIKE ? AND last LIKE ? AND addr LIKE ? AND zip LIKE ? AND date LIKE ?;"""
sql_insert = "INSERT INTO tOrderDetail (order_id, prod_id, qty) VALUES (?, ?, ?);"

for i, row in enumerate(tOrderDetail.to_dict(orient = 'records')):
    orderdetail_params = (f"%{row['first']}%", f"%{row['last']}%", f"%{row['addr']}%", f"%{row['zip']}%", f"%{row['date']}%") 
    curs.execute(sql_check, orderdetail_params)

    result = curs.fetchone()

    if result:
        order_id = result[0]
        try:
            curs.execute(sql_insert, (order_id, row['prod_id'], row['qty']))
        except Exception as e:
            print(e)
            print('\nrow: ', i)
            print('\n', row)
            break
```
